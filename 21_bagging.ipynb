```python
upper_limit = (df.RestingBP.mean() + 3 * df.RestingBP.std())
upper_limit
```

```python
df = df[(df.RestingBP >= lower_limit) & (df.RestingBP <= upper_limit)]
```

```python
upper_limit = (df.Cholesterol.mean() + 3 * df.Cholesterol.std())
upper_limit
```

```python
lower_limit = (df.Cholesterol.mean() - 3 * df.Cholesterol.std())
lower_limit
```

```python
df = df[(df.Cholesterol >= lower_limit) & (df.Cholesterol <= upper_limit)]
df
```

```python
upper_limit = (df.FastingBS.mean() + 3 * df.FastingBS.std())
lower_limit = (df.FastingBS.mean() - 3 * df.FastingBS.std())
df = df[(df.FastingBS >= lower_limit) & (df.FastingBS <= upper_limit)]
```

```python
upper_limit = (df.MaxHR.mean() + 3 * df.MaxHR.std())
lower_limit = (df.MaxHR.mean() - 3 * df.MaxHR.std())
df = df[(df.MaxHR >= lower_limit) & (df.MaxHR <= upper_limit)]
```

```python
upper_limit = (df.Oldpeak.mean() + 3 * df.Oldpeak.std())
lower_limit = (df.Oldpeak.mean() - 3 * df.Oldpeak.std())
df = df[(df.Oldpeak >= lower_limit) & (df.Oldpeak <= upper_limit)]
```

```python
df
```

```python
categorical_col = ['Sex', 'ChestPainType', 'RestingECG', 'ExerciseAngina', 'ST_Slope']
```

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
for column in categorical_col:
    df[column] = le.fit_transform(df[column])
```

```python
df
```

```python
X = df.drop('HeartDisease', axis = 'columns')
```

```python
y = df.HeartDisease
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.tree import DecisionTreeClassifier
```

```python
scores = cross_val_score(DecisionTreeClassifier(), X_scaled, y, cv = 10)
scores.mean()
```

```python
from sklearn.ensemble import BaggingClassifier
```

```python
bag_model = BaggingClassifier(
    estimator = DecisionTreeClassifier(), 
    n_estimators = 100, 
    max_samples = 0.8, 
    oob_score = True,
    random_state = 0
)
scores = cross_val_score(bag_model, X, y, cv=10)
scores.mean()
```
